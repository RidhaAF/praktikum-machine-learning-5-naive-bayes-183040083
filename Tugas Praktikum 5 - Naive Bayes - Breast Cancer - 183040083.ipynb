# Breast Cancer Wisconsin (diagnostic) dataset

```python
import numpy as np
import pandas as pd
```

```python
# Load datasets package from scikit-learn
from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
```

```python
print(data.keys())
```

```python
print(data.target_names)
```

```python
X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
X.isna().any()
```

```python
X.head()
```

# Preprocessing
### One-Hot Encoding

```python
X_enc = pd.get_dummies(X)
```

```python
X_enc.columns
```

```python
X = X_enc
y = data.target
```

```python
X
```

# Feature Selection
### Feature Selection: by Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold()
```

```python
XVarianceThreshold = selector.fit_transform(X)
```

```python
pd.DataFrame(XVarianceThreshold)
```

```python
X
```

```python
XVarianceThreshold.shape
```

### Feature Selection: by Mutual Information/Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
XMutualInformationValue = mutual_info_classif(X, y)
```

```python
XMutualInformationValue
```

```python
pd.DataFrame(XMutualInformationValue).median()
```

```python
XMutualInformation = X
```

```python
XMutualInformation
```

```python
# Mengambil nilai median mutual
XMutualInformation.loc[569] = XMutualInformationValue > 0.17
```

```python
XMutualInformation = XMutualInformation.drop([col for col in X.columns if X[col][569] == 0], axis = 1)
```

```python
XMutualInformation = XMutualInformation.drop(569)
```

```python
XMutualInformation = XMutualInformation.replace(to_replace = [0.0], value = 0)
XMutualInformation = XMutualInformation.replace(to_replace = [1.0], value = 1)
```

```python
XMutualInformation = XMutualInformation.astype(int)
```

```python
XMutualInformation
```

```python
X = X.astype(int)
X = X.drop(569)
```

# Melatih Model
### X hasil variance threshold (X)

```python
# Memisahkan dataset ke dalam set training dan set testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(XVarianceThreshold, y, test_size = 0.4, random_state = 1)
```

```python
X_train
```

```python
# Melatih model dengan set training
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# Membuat prediksi hasil set testing
predictions = gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from mlxtend.plotting import plot_confusion_matrix

myarray = np.asarray(y_test)
matrix = confusion_matrix((myarray), (predictions))

class_names = ['malignant', 'benign']

fig, ax = plot_confusion_matrix(conf_mat = matrix, show_absolute = True, show_normed = False, colorbar = True, class_names = class_names)
```
